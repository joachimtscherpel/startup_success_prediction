```python
def generate_leak_flag(text, model="mistral:7b-instruct-v0.3-q3_K_L"):
    if text is None or pd.isna(text) or str(text).lower() == 'nan':
        return None
    else:
        prompt = f"""In the following sentence, determine the value of the binary label "info" by
finding any concrete information on milestones, achievements, growth or success events such as funding, IPO, valuation, number of employees, acquisitions, sales figures, prominent clients or customers, expanding to new office locations, successful product launch and availability etc. Keywords are for example "fund", "round", "ipo", "acquisition", "bankruptcy", "merged", "expanded".

Rule:
If there is any of this information present in the text, set info: 1
If there is no information like that in the text, set info: 0.

START OF SENTENCE >>>
{text}
<<< END OF SENTENCE

Return the label with a short sentence of reasoning:
info: X - Reason: ...
Don't add an intruductory sentence.
"""

        seed = 4 #random.randint(1, 1000)
        # Choose zero temperature to get reliable and consistent results
        # High temperature leads to correct reasoning but wrong label
        result = llm.prompt(prompt, model, temp=0, seed=seed, num_predict=15, top_k=40, top_p=0.9)
        return result
```

```python
def generate_leak_flag_2(text, model="mistral:7b-instruct-v0.3-q3_K_L"):
    if text is None or pd.isna(text) or str(text).lower() == 'nan':
        return None
    else:
        prompt = f"""In the following sentence, determine the value of the binary label "info" by
finding any concrete information on failure events or turmoil such as bankrupty, ceased operations, inactivity, liquidation, lawsuits, ownership changes, out-of-business status, discontinuation, failure, shutdowns, operational challenges, scandals, exit, negative results, losses etc..

Rule:
If there is any of this information present in the text, set info: 1
If there is no information like that in the text, set info: 0.

START OF SENTENCE >>>
{text}
<<< END OF SENTENCE

Return the label with a short sentence of reasoning:
info: X - Reason: ...
Don't add an intruductory sentence.
"""

        seed = 4 #random.randint(1, 1000)
        # Choose zero temperature to get reliable and consistent results
        # High temperature leads to correct reasoning but wrong label
        result = llm.prompt(prompt, model, temp=0, seed=seed, num_predict=15, top_k=40, top_p=0.9)
        return result
```

```python
# Load the checkpoint file
with open('data/dr/text/leakage/dr_clean_2.pkl', 'rb') as f:
    data = pickle.load(f)

# Extract results from the 'results' key and convert to DataFrame
dr_clean = pd.DataFrame(data['results'])

dr_clean
```

```python
# Append investor and founder names for masking
dr = pd.read_pickle('data/dr/dr_preprocessed.pkl')
dr['id'] = dr['id'].astype(str)
dr_clean = dr_clean.merge(dr[["id", "round_investors", "founders"]], how="left", on="id")
dr_clean
```

```python
# Function to combine columns into lists
def combine_to_list(row):
    # flatten round_investors safely
    investors = row.get('round_investors', [])
    if isinstance(investors, list):
        investors_flat = list(chain.from_iterable([i for i in investors if isinstance(i, list)]))
    else:
        investors_flat = []

    # founders
    founders = row.get('founders', [])
    if not isinstance(founders, list):
        founders = []

    # name
    name = [row.get('name')] if pd.notna(row.get('name')) else []

    # Return as tuple
    return pd.Series([name + founders, investors_flat])

# Apply and create two new columns
dr_clean[['names', 'investor_names']] = dr_clean.apply(combine_to_list, axis=1)

dr_clean.head()
```

```python
dr_clean['masked'] = dr_clean.apply(lambda row: llm.spacy_mask(row['clean_2'], row['names'], row['investor_names']), axis=1)
dr_clean
```

```python
dr_clean.to_pickle('data/dr/text/leakage/dr_masked.pkl')
```
